### pymupdf

I mean it gets all the content without watermarking but the formatting is kinda shite

```python
# now i'll try to use pymupdf instead
import fitz
from pathlib import Path
import os

root_dir = Path(os.getcwd()).parent
data_dir = root_dir / "data"

filename = "Essay - Managed by AI not replaced.pdf"
pdf_path = data_dir / "academic" / filename

doc = fitz.open(pdf_path)
html_content = []

for page in doc:
    html_content.append(page.get_text("xhtml"))

output_path = data_dir / filename.replace('.pdf', '.html')
with open(output_path, 'w', encoding='utf-8') as f:
    f.write('\n'.join(html_content))

doc.close()
print(f"Conversion complete: {output_path}")
```

### okay pdfplumber

```python
import pdfplumber
from pathlib import Path
import os

root_dir = Path(os.getcwd()).parent
data_dir = root_dir / "data"

filename = "Essay - Managed by AI not replaced.pdf"
pdf_path = data_dir / "academic" / filename

html_content = ['<html><head><meta charset="UTF-8"></head><body>']

with pdfplumber.open(pdf_path) as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        # Wrap each page in a div
        html_content.append(f'<div class="page"><p>{text.replace(chr(10), "<br>")}</p></div>')

html_content.append('</body></html>')

output_path = data_dir / filename.replace('.pdf', '.html')
with open(output_path, 'w', encoding='utf-8') as f:
    f.write('\n'.join(html_content))

print(f"Conversion complete: {output_path}")
```
